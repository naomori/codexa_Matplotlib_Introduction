# Question.2

```
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y1 = x*2
y2 = x**2

plt.scatter(x, y1, label="x*2")
plt.scatter(x, y2, label="x**2")
plt.legend()
plt.show()
```

# Question.4

```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 15)

y1 = x*3 + 10
y2 = x*(-5)-20
y3 = x**2 - 5

plt.plot(x, y1, "rx")
plt.plot(x, y2, "go")
plt.plot(x, y3, "b-")
plt.show()
```

# Question.5

```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 20)

y1 = np.sin(x)
y2 = 12*x + 8
y3 = x**2
y4 = np.exp(x)

figure, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

axes[0,0].plot(x, y1,"r")
axes[0,1].plot(x, y2,"b")
axes[1,0].plot(x, y3,"g")
axes[1,1].plot(x, y4,"-")
figure.show()
```
